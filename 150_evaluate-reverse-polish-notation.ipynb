```python
from operator import add,sub,mul
def div(x,y):
        return int(x/y)
class Solution:
    eval={'+':add,'-':sub,'*':mul,'/':div}
    def evalRPN(self, tokens: List[str]) -> int:
        stack=[]
        for char in tokens:
            if char in ['+','-','*','/']:
                temp1=stack.pop()
                temp2=stack.pop()
                stack.append(self.eval[char](temp2,temp1))
            else:
                stack.append(int(char))
        return stack[0]
```
